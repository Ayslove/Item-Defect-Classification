```python
import pandas as pd
import pickle
```

```python
# load the model
with open('model.pkl', 'rb') as file_1:
    model = pickle.load(file_1)
```

```python
# Create Dummy Data
data_dummy = {
    'MaintenanceHours': [1, 2, 3],
    'ProductionVolume': [200, 126, 256],
    'DefectRate': [3.7, 1.2, 0.5],
    'QualityScore': [82.6, 77.2, 85.3],
    'ProductionCost': [5000, 7000, 4500],
    'SupplierQuality': [90.5, 85.2, 88.7],
    'DeliveryDelay': [1, 2, 0],
    'DowntimePercentage': [1.2, 2.5, 0.8],
    'InventoryTurnover': [5.6, 4.2, 6.1],
    'StockoutRate': [2.3, 3.1, 1.8],
    'WorkerProductivity': [92.5, 89.3, 95.1],
    'SafetyIncidents': [1, 0, 2],
    'EnergyConsumption': [3000, 3500, 2800],
    'EnergyEfficiency': [0.3, 0.25, 0.35],
    'AdditiveProcessTime': [5.5, 6.2, 4.8],
    'AdditiveMaterialCost': [200.0, 180.0, 210.0]
}

# Create DataFrame
df_data_dummy = pd.DataFrame(data_dummy)

# Display DataFrame
df_data_dummy.head()
```

```python
# Predict the defect items
predicted_deffect = model.predict(df_data_dummy)
predicted_deffect
```

```python
# assign prediction to the data
df_data_dummy['prediction'] = predicted_deffect
df_data_dummy['prediction'] = df_data_dummy['prediction'].map({1: 'Defect', 0: 'Not Defect'})

# return readable prediction data from user
df_data_dummy[['MaintenanceHours','ProductionVolume', 'DefectRate', 'QualityScore','prediction']]
```
